```python
import pandas as pd
import numpy as np
```

```python
# import accident data (numeric values)
accidents = pd.read_csv('accidents_Cleaned.csv')
```

```python
# create a series of dictionaries
PCHCTT = {0:'None within 50 metres', 
          1:'Control by school crossing patrol',
          2:'Control by other authorised person'};

PCPFTT = {0:'No physical crossing within 50 meters',
          1:'Zebra crossing', 
          4:'non-junction pedestrian crossing',
          5:'Pedestrian phase at traffic signal junction',  
          7:'Footbridge or subway',
          8:'Central refuge'};

WCTT = {1:'Fine without high winds', 
        2:'Raining without high winds', 
        3:'Snowing without high wind',    
        4:'Fine with high winds', 
        5:'Raining with high winds', 
        6:'Snowing with high winds',
        7:'Fog or mist', 
        8:'Other',
        9:'Unknown'};

RSCTT = {1:'Dry', 
         2:'Wet/Damp', 
         3:'Snow', 
         4:'Frost/Ice', 
         5:'Flood (Over 3cm of water)'};

SCASTT = {0:'None',
          1:'Auto traffic singal out',
          2:'Auto signal partly defective', 
          3:'Road sign or marking defective or obscured', 
          4:'Roadworks',
          5:'Road surface defective',
          6:'Oil or diesel', 
          7:'Mud'};

CHTT = {0:'None', 
        1:'Vehicle load on road',
        2:'Other object on road',
        3:'Previous accident',
        6:'Pedestrian in carriageway (not injured)',    
        7:'Any animal (except a ridden horse)'};

OATT = {1:'Yes', 2:'No', 
        3: 'No - accident was reported using a self completion form'};

RTTT = {1:'Roundabout',
        2:'One way street', 
        3:'Dual carriageway', 
        6:'Single carriageway',
        7:'Slip road', 
        9:'Unknown'};

LCTT = {1:'Daylight',
        4:'Darkness-lights lit',
        6:'Darkness-no lighting',
        5:'Darkness-lights unlit',
        7:'Darkeness-lighting unknown'};

ASTT = {1:'Fatal', 2:'Serious', 3:'Slight'}

UORTT = {1:'Urban', 2:'Rural', 3:'Unallocated'}

SODTT = {1:'Male', 2:'Female', 3:'Not known'}

ABODTT = {1:'0-5',
          2:'6-10',
          3:'11-15',
          4:'16-20',
          5:'21-25',
          6:'26-35',
          7:'36-45',
          8:'46-55',
          9:'56-65',
          10:'66-75',
          11:'over 75'};

DOWTT = {1:'Sunday',
         2:'Monday',
         3:'Tuesday',
         4:'Wednesday',
         5:'Thursday',
         6:'Friday',
         7:'Saturday'};
```

```python
# convert the numeric values to categoric string values
accidents['Pedestrian_Crossing-Human_Control'] = accidents['Pedestrian_Crossing-Human_Control'].map(PCHCTT);
accidents['Pedestrian_Crossing-Physical_Facilities'] = accidents['Pedestrian_Crossing-Physical_Facilities'].map(PCPFTT);
accidents['Weather_Conditions'] = accidents['Weather_Conditions'].map(WCTT);
accidents['Road_Surface_Conditions'] = accidents['Road_Surface_Conditions'].map(RSCTT);
accidents['Special_Conditions_at_Site'] = accidents['Special_Conditions_at_Site'].map(SCASTT);
accidents['Carriageway_Hazards'] = accidents['Carriageway_Hazards'].map(CHTT);
accidents['Did_Police_Officer_Attend_Scene_of_Accident'] = accidents['Did_Police_Officer_Attend_Scene_of_Accident'].map(OATT);
accidents['Road_Type'] = accidents['Road_Type'].map(RTTT);
accidents['Light_Conditions'] = accidents['Light_Conditions'].map(LCTT);
accidents['Accident_Severity'] = accidents['Accident_Severity'].map(ASTT);
accidents['Urban_or_Rural_Area'] = accidents['Urban_or_Rural_Area'].map(UORTT);
accidents['Sex_of_Driver'] = accidents['Sex_of_Driver'].map(SODTT);
accidents['Age_Band_of_Driver'] = accidents['Age_Band_of_Driver'].map(ABODTT);
accidents['Day_of_Week'] = accidents['Day_of_Week'].map(DOWTT);
```

```python
# convert some of the float64 data types to int64 data types
accidents['Number_of_Vehicles'] = accidents['Number_of_Vehicles'].astype('int64');
accidents['Number_of_Casualties'] = accidents['Number_of_Casualties'].astype('int64');
accidents['Local_Authority_(District)'] = accidents['Local_Authority_(District)'].astype('int64');
accidents['Age_of_Driver'] = accidents['Age_of_Driver'].astype('int64');
accidents['Age_of_Vehicle'] = accidents['Age_of_Vehicle'].astype('int64');
accidents['Speed_limit'] = accidents['Speed_limit'].astype('int64')
```
