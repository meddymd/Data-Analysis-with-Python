```python
import pandas as pd
import numpy as np
```

```python
# import data from csv file
accidents = pd.read_csv('Accidents0515.csv',index_col='Accident_Index')
casualties = pd.read_csv('Casualties0515.csv' , error_bad_lines=False,index_col='Accident_Index',warn_bad_lines=False)
vehicles = pd.read_csv('Vehicles0515.csv', error_bad_lines=False,index_col='Accident_Index',warn_bad_lines=False)
```

```python
# overview of the first 5 rows
accidents.head(5)
```

```python
# data types of all the columns 
# the discription of the columns are in the file "7752_road-accident-safety-data-guide.xls"
accidents.dtypes
```

```python
# check completeness of the data and row number
accidents
```

```python
# identifies duplicate rows in the data 
accidents.duplicated(subset=None, keep=False).sum()
```

```python
# drop duplicates
# in reality, accident events are very unlikely to have exactly the same characteristics such as date, time, location and number of vehicles and thus should be removed from the dataset.
accidents.drop_duplicates(subset=None, keep=False, inplace=True) 
```

```python
# join the accident dataset with the vehicles dataset
accidents = accidents.join(vehicles, how='outer')
```

```python
# check the data type of all the columns again after the join with vehicle data
accidents.dtypes
```

```python
# identifies all null values in the dataset.
accidents.isnull().sum()
```

```python
# The value â€œ-1â€Â is also exported for NULLÂ or out of range values.
np.sum(accidents == -1)
```

```python
# drop unwanted columns
accidents.drop('1st_Road_Class', axis =1, inplace=True);
accidents.drop('1st_Road_Number', axis =1, inplace=True);
accidents.drop('2nd_Road_Class', axis =1, inplace=True);
accidents.drop('2nd_Road_Number', axis =1, inplace=True);
accidents.drop('Location_Easting_OSGR', axis=1, inplace=True);
accidents.drop('Location_Northing_OSGR', axis=1, inplace=True);
accidents.drop('Junction_Control', axis =1, inplace=True);
accidents.drop('LSOA_of_Accident_Location', axis =1, inplace=True);
```

```python
# create column Date_Time
accidents['Date_Time'] =  accidents['Date'] +' '+ accidents['Time']
```

```python
# drop unwanted rows (the rows with missing values)
for col in accidents.columns:
    accidents = (accidents[accidents[col]!=-1])

for col in casualties.columns:
    casualties = (casualties[casualties[col]!=-1])

    accidents['Date_Time'] = pd.to_datetime(accidents.Date_Time)

# drop column Date and Time, since column Date_Time is created
accidents.drop(['Date','Time'],axis =1 , inplace=True)
accidents.dropna(inplace=True)
```

```python
# check if there are still invalid values in the dataset (NULL or -1)
accidents.isnull().sum()
```

```python
np.sum(accidents==-1)
```

```python
# create backup and exporte the cleaned dataset to a file
accidents_backup = accidents.copy(deep=True)
accidents.to_csv('accidents_Cleaned.csv') 
```
